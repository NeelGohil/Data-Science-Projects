```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
%matplotlib inline
```

```python
file = open('university.data', 'r').read()
```

```python
ls = []
for univ in file.split('))'):
    #print(univ)
    for entry in univ.split('('):
        entry = entry.strip("'\n',) ''")
        entry.rsplit(' ', maxsplit=1)
        
        ls.append(entry)
        
        
                       
cleaned_list = list(filter(None, ls))
```

```python
cleaned_list
```

```python
splitted = []
for i in cleaned_list:
    i = i.split()
    splitted.append(i)
    
    
```

```python
values = []
for i in splitted:
    values.append(i[-1])
```

```python
keys = []
for j in splitted:
    keys.append(j[:-1])
    
```

```python
df1 = pd.Series(keys)
df2 = pd.Series(values)
```

```python
df= pd.concat([keys,values], axis = 1)
```
