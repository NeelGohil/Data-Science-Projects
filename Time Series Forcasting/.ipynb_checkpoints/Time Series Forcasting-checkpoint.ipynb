```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
births = pd.read_csv('daily-total-female-births-CA.csv', index_col= 'date', parse_dates= [0])
series_values = births.values
```

```python
# Exploring data set

#births.head()
births.describe()
```

```python
#Quick Visualization to understand births per month

births.plot()
```

```python
#Creating moving average to smoothen the chart

births_12 = births.rolling(window = 30).mean()
```

```python
births_12.plot()
```

```python
value = pd.DataFrame(series_values)
```

```python
# Creating baseline model of timeseries
birth_df = pd.concat([value, value.shift(1)], axis = 1)
```

```python
birth_df
```

```python
birth_df.columns = ['Actual_births', 'Forcasted_birth']
```

```python
birth_df
```

```python
from sklearn.metrics import mean_squared_error
birth_test = birth_df[1:]
birth_test.head()
birth_error = mean_squared_error(birth_test['Actual_births'], birth_test['Forcasted_birth'])
```

```python
np.sqrt(birth_error)
```
